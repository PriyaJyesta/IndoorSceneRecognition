```python
extensions = ("*.jpg")

image_arrays = []
labels = []

for scene in indoor_Scene_list:
    for ext in extensions:
        for index, pic in enumerate(glob.glob(pic_lib + "/" + scene + "/" + ext), start = 1):
            #file name to label           
            labels.append(scene)
           
            #split file path and file name 
            path, filename = os.path.split(pic)
            
            #New array Name
            newFileName = str(scene)+str(index)
            
            img = cv2.imread(os.path.join(pic_lib, pic), 0)  
            
            #new_img_array = cv2.resize(img, dsize=(80, 80))
            image_arrays.append(img)
            
```
